```python
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
dataFrame=pd.read_csv('ds_salaries.csv')
dataFrame.head()
```

```python
salarios = dataFrame['salary']
profissoes = dataFrame['job_title']
```

```python
plt.scatter(salarios, profissoes)
plt.show()
```

```python
plt.plot(profissoes, salarios)
plt.show()
```

```python
px.line(x=profissoes, y=salarios)
```

```python
#os dois grficos gerados a partir das colunas de salrios e cargo precisam ser limpos
# e o grfico feito a partir das mesmas colunas com os ploty.express mostra a maior
# parte da faixa salarial abaxio de 5 milhoes, alguns numa faixa media de at 11 milhes
# e apenas um tipo de cargo recebendo mais do isso, passando at de 30 milhes, porm numa
# tem uma relao desse aumento e o tipo da moeda no qual o salrio  pago.
```
