```python
import requests as r
```

```python
url = 'https://api.covid19api.com/dayone/country/brazil'
resp = r.get(url)
```

```python
resp.status_code
```

```python
raw_data = resp.json()
```

```python
raw_data[0]
```

```python
data =[]
for obs in raw_data:
    data.append([obs['Confirmed'], obs['Deaths'], obs['Recovered'], obs['Active'], obs['Date']])
```

```python
data.insert(0, ['confirmados', 'obitos', 'recuperados', 'ativos', 'data'])
data
```

```python
CONFIRMADOS=0
OBITOS=1
RECUPERADOS=2
ATIVOS=3
DATA=4
```

```python
for i in range(1, len(data)):
    data[i][DATA] = data[i][DATA][:10] 
```

```python
#Quickchart documentation
def get_datasets(y, labels):
    if type(y[0]) == list:
        datasets=[]
        for i in range(len(y)):
            datasets.append({
                'label': labels[i],
                'data': y[i]
            })
        return datasets
    else:
        return[
            {
                'label': labels[0],
                'data': y
            }
        ]
```

```python
def set_title(title=''):
    if title != '':
        display='true'
    else:
        display='false'
    return{
        'title':title,
        'display': display
    }
```

```python
def create_chart(x,y,labels,kind='bar',title=''):
    datasets = get_datasets(y, labels)
    options = set_title(title)
    
    chart = {
        'type': kind,
        'data':{
            'labels': x,
            'datasets': datasets
        },
        'options':options
    }
    return chart
```

```python
def get_api_chart(chart):
    url_base = 'https://quickchart.io/chart'
    resp = r.get(f'{url_base}?c={str(chart)}')
    return resp.content
```

```python
def save_image(path, content):
    with open (path, 'wb') as image:
        image.write(content)
```

```python
from PIL import Image
from IPython.display import display
```

```python
def display_image(path):
    img_pill = Image.open(path)
    display(img_pill)
```

```python
y_data=[]
for obs in data[1:10]:
    y_data.append(obs[CONFIRMADOS])
    
y_data2 =[]
for obs in data[1:10]:
    y_data2.append(obs[CONFIRMADOS])
labels = ['Confirmado', 'Recuperados']

x= []

for obs in data[1:10]:
    x.append(obs[DATA].strftime('%d%m%Y'))

chart = create_chart(x, [y_data, y_data2], labels, title='Grafico confirmado vs recuperados')
chart_content = get_api_chart(chart)
save_image('meu-primeiro-grafico.png', chart_content)
display_image('meu-primeiro-grafico.png')
```

```python
from urllib.parse import quote
def get_api_qrcode(link):
    text = quote(link)
    url_base = 'https://quickchart.io/qr'
    resp = r.get(f'{url_base}?text={text}')
    return resp.content
```

```python
url_base = 'https://quickchart.io/chart'
link = f'{url_base}?c={str(chart)}'
save_image('qrcode.png', get_api_qrcode(link))
display_image('qrcode.png')
```
